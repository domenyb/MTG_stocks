```python
import pandas as pd
from glob2 import glob
```

First I'll go through the individual sets and pack them into one csv file per set for easier handling. 

This is done by loading my naming convention from the editions.dat, and merging every csv into one dataframe. 
I'll compare the ones following each other in order to remove duplicates.

I'll remove all cards that doesn't really carry information in the long run. Most common (and/or uncommon) cards have no real value, but I don't want to use exclusively rare cards for the project. The next step would be to look at the distribution of prices and determine a lower limit, under which I'd include all the cards, but I'll skip this step and introduce a uniform 0.2 $ limit from experience. 

So if a card didnt have (nor did its foil counterpart) over 0.2\\$ (correction as of 22.07.2020: this is still waay too low, I would like to go up to 0.4\\$. If that is still not sufficient, I'll require the mean value of a card to be above 0.25\\$) at any given time over the period, then I'll exclude that card.

I'm aware that this removes more than half of my data, but that is not really valuable data since those cards are essentially worthless and the playerbase has a huge overabundance of them. 


```python
set_names=pd.read_csv("editions.dat",sep=";", names=["URL", "set_name"])
#Load the file which contains the set names

list_of_csvs=glob("csv/"+set_names.set_name[0]+"_*")
df=pd.read_csv(list_of_csvs[0])
df.keys()
#let's decide which attributes are the ones I'd actually need. I'm aware of some editions having extra info, but those columns are irrelevant as well. I just intend to create a list of the useful columns
df=df.drop(labels=df.keys()[:2],axis=1)# I don't trust these values, and will calculate it anyway when needed - so for now I don't remove information with it
df=df.drop(labels=df.keys()[2],axis=1)#The card name is sufficient to identify cards
df=df.drop(labels=df.keys()[4:6],axis=1)# see the comment above
df=df.drop(labels=df.keys()[-6:-4],axis=1)# once again, no information lost with these
df=df.drop(labels=df.keys()[-2:],axis=1)
key_list=df.keys()
```

```python




for item in range(0,len(set_names.set_name),2):# I need to skip every second here - since I load foil & nonfoil together
    buff_prev=pd.DataFrame()#this is always the previous item for comparison
   # list_of_csvs=glob("csv/"+item+"*")
    edition=[pd.DataFrame(), pd.DataFrame()]
    for foil in range(2):
        list_of_csvs=glob("csv/"+set_names.set_name[item+foil]+"_*")
        for datapoint in list_of_csvs:
            current=pd.read_csv(datapoint)
            if not buff_prev.equals(current):# so if there is at least one change, we continue
                buff_prev=current.copy() # I have to save the current as a new copy before any alteration
                current=current[key_list]
                current["date"]=datapoint.split("_")[1]+"_"+datapoint.split("_")[2].split(".")[0] #I could've chosen a better format for the naming convention, but oh well
                # I've added the datapoint, now give it to the full
                edition[foil]=pd.concat([edition[foil], current], ignore_index=True, sort=True)
    ##### Now both the foil and non-foil "dummies" are made, exclude the cards with no value
    print ("Finished concatenating "+set_names.set_name[item]+", number of unique cards: "+str(len(edition[0].name.unique())))
    for card_name in edition[0].name.unique():
        if edition[0][edition[0].name==card_name].fair_price.max()<.4 and edition[1][edition[1].name==card_name].fair_price.max()<.4:#sometimes I regret using these long names but they are descriptive...
            print("EXCLUDED "+card_name )
            edition[0]=edition[0][edition[0].name!=card_name]
            edition[1]=edition[1][edition[1].name!=card_name]            
    #write them into a new csv - probably not necessary, but I like to have my work saved, for now I handle my storage space as infinite with a note that this is not always the case.
 #   edition[0].to_csv(set_names.set_name[item]+"_unified.csv",index=False)
 #   edition[1].to_csv(set_names.set_name[item]+"F_unified.csv",index=False)
    
```
